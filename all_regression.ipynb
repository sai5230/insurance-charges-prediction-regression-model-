```python
#finding outliers in data
plt.figure(figsize = (6,4))
sns.boxplot(x = df["age"])
plt.title("boxplot")
plt.show()
```

```python
#in this there is no outliers but we try ti apply iqr to remove outliers
Q1 = df["age"].quantile(0.25)
Q3 = df["age"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print("Outlier range: <", lower_bound, "or >", upper_bound)
```

```python
df_no_outliers = df[
    (df["age"] >= lower_bound) & (df["age"] <= upper_bound)
]

print("After removing outliers:", df_no_outliers.shape)
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df_no_outliers["age"])
plt.title("Boxplot")
plt.show()
```

```python
x = df.drop("charges",axis = 1)
y = df["charges"]
x
```

```python
y
```

```python
#scaling of data
"""used for makiing the data into a range
"""
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

```python
#splitting into taining and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
#using linear regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error,root_mean_squared_error
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
a = r2_score(y_test,y_pred)*100
b = mean_squared_error(y_test,y_pred)
c = root_mean_squared_error(y_test,y_pred)
print(a)
print(b)
print(c)
```

```python
#using descision tree
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score,mean_squared_error,root_mean_squared_error
dr = DecisionTreeRegressor()
dr.fit(x_train,y_train)
y_pred = dr.predict(x_test)
a = r2_score(y_test,y_pred)*100
b = mean_squared_error(y_test,y_pred)
c = root_mean_squared_error(y_test,y_pred)
print(a)
print(b)
print(c)
```

```python
#random forest
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
a = r2_score(y_test,y_pred)*100
b = mean_squared_error(y_test,y_pred)
c = root_mean_squared_error(y_test,y_pred)
print(a)
print(b)
print(c)
```

```python
#svr
from sklearn.svm import SVR
sv = SVR()
sv.fit(x_train,y_train)
y_pred = sv.predict(x_test)
a = r2_score(y_test,y_pred)*100
b = mean_squared_error(y_test,y_pred)
c = root_mean_squared_error(y_test,y_pred)
print(a)
print(b)
print(c)
```
